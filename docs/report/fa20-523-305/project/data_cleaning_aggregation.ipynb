```python
import os
import pandas as pd
```

# Aggregate All Soil Moisture Files

```python
#Move content of all files 
file_list = os.listdir('caf_sensors/Daily/')
agg_data = pd.DataFrame()
for file in file_list:
    path = 'caf_sensors/Daily/' + file
    curr_data = pd.read_csv(path, sep='\t')
    agg_data = agg_data.append(curr_data)
```

```python
# Get percent of NAs in each column
(len(agg_data) - agg_data.count())/len(agg_data)
```

```python
#Remove rows that only have NAs
agg_data = agg_data.dropna(thresh=10)
```
